```python
from astropy.table import Table

# convert pandas DataFrame to Astropy Table with object columns as arrays
def pandas_to_astropy_table(df):
    table = Table()
    for col in df.columns:
        if df[col].dtype == object:
            # Convert string representation of arrays back to actual arrays
            def to_array(x):
                if isinstance(x, str):
                    return np.fromstring(x.strip("[]"), sep=",")
                else:
                    return np.array([])

            array_col = df[col].apply(to_array).tolist()
            table[col] = array_col
        else:
            table[col] = df[col].values
    return table

astropy_table = pandas_to_astropy_table(df)
```
