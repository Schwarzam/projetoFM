```python
from astromodal.tokenizers import SpectrumTokenizer

groups = [
    {"name": "gaiaxp_bp", "kind": "gaiaxpy_xp", "xp_band": "BP"},
    {"name": "gaiaxp_rp", "kind": "gaiaxpy_xp", "xp_band": "RP"},
    {"name": "desi_b", "kind": "flux_ivar",
    "flux_col": "desi_flux_b", "ivar_col": "desi_ivar_b"},
    # ... more groups
]

tokenizer = SpectrumTokenizer(
    codebook_size=2048,
    max_features_per_group=2_000_000,
    max_rows_per_file=5000,
    max_points_per_row=2000,
    groups=groups,
    gaiaxpy_mode="convert",
    gaiaxpy_truncation=False,
)

import glob

parquet_files = glob.glob("/home/astrodados4/downloads/hypercube/datacube_*.parquet")
tokenizer.train(parquet_files)
```
